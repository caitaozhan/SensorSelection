```python
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
def_figsize = 60, 45

SMALL_SIZE =20
MEDIUM_SIZE=20
BIGGER_SIZE=20

plt.style.use('seaborn-paper')
sns.set(context='paper', font_scale=1)
sns.set_style("whitegrid")

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) # fontsize of the figure title

df = pd.read_csv('../../plot_data_splat/splot-output.txt', header=None, delimiter=' ')
df2 = pd.read_csv('../../plot_data_splat/clustering-output.txt', header=None, delimiter=' ')
#df.columns = ['Error', 'Misses']
clustering_mean_error = np.mean(df2.loc[:, 0])
clustering_mean_misses = np.mean(df2.loc[:, 1] / 20)
clustering_mean_false = np.mean(df2.loc[:, 2] / 20)

splot_mean_error = np.mean(df.loc[:, 0])
splot_mean_misses = np.mean(df.loc[:, 1] / 20)
splot_mean_false = np.mean(df.loc[:, 2] / 20)
#print(mean_misses, mean_false)
                        
#df = pd.read_csv('fig1-homo/GA', header=None)
#X2, Y2 = df[0].tolist(), df[1].tolist()         # Y2: GA

#df = pd.read_csv('fig1-homo/coverage', header=None)
#X3, Y3 = df[0].tolist(), df[1].tolist()         # Y3: coverage

#df = pd.read_csv('fig1-homo/random', header=None)
#X4, Y4 = df[0].tolist(), df[1].tolist()         # Y4: random

plt.bar(range(0, 3), [0, clustering_mean_error , splot_mean_error * 12.5])
plt.xticks(range(0, 3), ['Our Algorithm', 'Clustering', 'SPLOT'])
plt.xlabel('Localization Algorithm')
plt.ylabel('Mean Localization Error (meters)')
plt.savefig('error.png')
plt.show()

plt.bar(range(0, 3), [0, clustering_mean_misses, splot_mean_misses])
plt.xticks(range(0, 3), ['Our Algorithm', 'Clustering', 'SPLOT'])
plt.xlabel('Localization Algorithm')
plt.ylabel('Percentage of Misses (%)')
plt.savefig('misses.png')
plt.show()

plt.bar(range(0, 3), [0, clustering_mean_false, splot_mean_false])
plt.xticks(range(0, 3), ['Our Algorithm', 'Clustering', 'SPLOT'])
plt.xlabel('Localization Algorithm')
plt.ylabel('Percentage of False Alarms (%)')
plt.savefig('fa.png')
plt.show()
#X1.insert(0, 0), Y1.insert(0, 0)
#X2.insert(0, 0), Y2.insert(0, 0)
#X3.insert(0, 0), Y3.insert(0, 0)
#X4.insert(0, 0), Y4.insert(0, 0)
```
